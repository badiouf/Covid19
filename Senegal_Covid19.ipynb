```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import folium
import seaborn as sns
sns.set(rc={'figure.figsize': (20, 10)})
import warnings
warnings.simplefilter('ignore')
```

```python
data=pd.read_csv('COVID_Dakar.csv',sep=';')
```

```python
data.head()
```

```python
from geopy.geocoders import Nominatim
geolocator = Nominatim()
data['latitude']='NaN'
data['longitude']='NaN'
data1=data.set_index('Ville')
for i in list(data['Ville'].dropna().unique()):
    location = geolocator.geocode(str(i),timeout=10000)
    data1.loc[[str(i)],['latitude']],data1.loc[[str(i)],['longitude']]=location.latitude,location.longitude
data1=data1.reset_index()
```

```python
data1.head()
```

```python
data1['Date'] = pd.to_datetime(data1['Date'], dayfirst=True)
data1.info()
```

# Statistiques descriptives

```python
data1['nb_test']=1
stat_pos_test_cum=data1[['Date','Positif','nb_test','Décédé','Guéri']].groupby('Date').sum().cumsum().reset_index()
stat_pos_test_cum['taux de positifs']=stat_pos_test_cum['Positif']/stat_pos_test_cum['nb_test']
stat_pos_test_cum['Evolution cas positifs']=0

for i in range(stat_pos_test_cum.shape[0]):
    if i==0:
        stat_pos_test_cum['Evolution cas positifs'].iloc[i]=0
    else:
        stat_pos_test_cum['Evolution cas positifs'].iloc[i]=stat_pos_test_cum['Positif'].iloc[i]/stat_pos_test_cum['Positif'].iloc[i-1]-1
```

```python
stat_pos_test_cum.head()
```

## Stat macro

```python
plt.figure(figsize=(20,10))
plt.plot(stat_pos_test_cum['Date'],stat_pos_test_cum['Positif'])
plt.legend()
plt.title('Evolution des cas testés positifs au Coronavirus: Sénégal')
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.plot(stat_pos_test_cum['Date'],stat_pos_test_cum['Evolution cas positifs'])
plt.legend()
plt.title('Croissance des cas testés positifs au Coronavirus: Sénégal')
plt.show()
```

# Commentaire:

Nous constatons que le taux d'évolution des cas positifs oscille de manière importante. Ce qui témoigne de la volatilité du nombre de cas positifs. Au 21 mars le taux d'volution journalier est de 22%

 ## Statistiques par type de patient

```python
stat_pos_test_client=stat_pos_test_cum[['Date','Positif','nb_test','Décédé','Guéri']]
```

```python
plt.figure(figsize=(20,8))
ax = stat_pos_test_client.groupby("Date").sum().plot(kind='area')
ax.set_ylabel('Effectif')
ax.set_title("Statistiques macro Coronavirus Sénégal")
```

```python
plt.figure(figsize=(20,8))
plt.plot(stat_pos_test_cum['Date'],stat_pos_test_cum['Positif'],label ='Nbre de cas positifs')
plt.plot(stat_pos_test_cum['Date'],stat_pos_test_cum['nb_test'],label ='Nbre de malades testés')
plt.plot(stat_pos_test_cum['Date'],stat_pos_test_cum['Guéri'],label ='Nbre de cas guéris')
plt.plot(stat_pos_test_cum['Date'],stat_pos_test_cum['Décédé'],label ='Nbre de décés')
plt.title('Statistiques macro Coronavirus Sénégal')
plt.legend()
plt.show()
```

# Commentaire:

Au 21 mars 2020: Sur les 237 cas testés, 56 cas sont confirmés positifs dont 5 guéris et 0 décédé


# Visualisation

```python
stat_ville=data1[['Ville','Positif','nb_test','Décédé','Guéri']].groupby('Ville').sum().cumsum().reset_index()
location=data1[['Ville','latitude','longitude']]
stat_ville=stat_ville.merge(location, left_on="Ville",right_on="Ville")
```

```python
stat_ville=stat_ville.drop_duplicates()
```

```python
stat_ville
```

```python
colordict = {0: 'lightblue', 18: 'lightgreen', 19: 'orange', 31: 'red'}
corona_map = folium.Map(location=[14, -15], zoom_start=5)
for lat, lon, Deces, Positif, city in zip(stat_ville['latitude'], stat_ville['longitude'], stat_ville['Décédé'], stat_ville['Positif'], stat_ville['Ville']):
    folium.CircleMarker(
        [lat, lon],
        radius=Positif*0.5,
        popup = ('Ville: ' + str(city).capitalize() + '<br>'
                 'Nbre cas positifs: ' + str(Positif) + '<br>'
                 'Taux de deces: ' + str(Deces/Positif) +'%'
                ),
        color='b',
        fill_color=colordict[Positif],
        fill=True,
        fill_opacity=0.7
        ).add_to(corona_map)
    
corona_map
```

```python
corona_map.save('map_corona.html')
```
